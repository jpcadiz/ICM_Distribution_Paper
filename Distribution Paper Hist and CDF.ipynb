```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Path to your CSV file
csv_file_path = 'Distribution Paper - Low Demand 500.csv'  # Replace with your actual file path

# Reading the CSV file into a DataFrame, skipping the first row and removing unnamed columns
df = pd.read_csv(csv_file_path, skiprows=1)

# Dropping 'Unnamed' columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Assigning new column names
df.columns = [
    "New Haven Gas Trucks", "New Haven Liquid Trucks", "New Haven Pipelines",
    "Bakersfield Gas Trucks", "Bakersfield Liquid Trucks", "Bakersfield Pipelines",
    "Atlanta Gas Trucks", "Atlanta Liquid Trucks", "Atlanta Pipelines",
    "Washington DC Gas Trucks", "Washington DC Liquid Trucks", "Washington DC Pipelines",
]

# Desired bin width
bin_width = 0.1

# Function to plot histogram as proportions with specific bin width
def plot_histogram_proportions(data, ax, label, bin_width):
    # Calculate the range of the data
    data_range = data.max() - data.min()
    # Determine the number of bins based on the range and desired bin width
    num_bins = int(data_range / bin_width)
    # Plot histogram with the calculated number of bins
    counts, bins, patches = ax.hist(data, bins=num_bins, alpha=0.7, label=f"{method} (bin width = {bin_width})", weights=np.ones_like(data) / len(data))
    for patch in patches:
        patch.set_edgecolor('black')  # Add black outline to bars
    #ax.set_ylim(0, max(counts) * 1.1)  # Adjust the y-axis limit for better visual spacing

# Plotting histograms as proportions
fig, axs = plt.subplots(4, 1, figsize=(15, 15))

cities = ["New Haven", "Bakersfield", "Atlanta", "Washington DC"]
methods = ["Gas Trucks", "Liquid Trucks", "Pipelines"]

for i, city in enumerate(cities):
    for method in methods:
        column_name = f"{city} {method}"
        if column_name in df.columns:
            plot_histogram_proportions(df[column_name], axs[i], method, bin_width)
    axs[i].set_title(f'{city}', fontsize = 25, fontweight = 'bold')
    axs[i].set_xlabel('Levelized Cost of Hydrogen ($/kg)', fontsize = 18)
    axs[i].set_ylabel('Proportion', fontsize = 18)
    axs[i].legend()
    axs[i].grid(True)

plt.tight_layout()

# Save the figure as an image file
plt.savefig('Low_Demand_Histograms')
plt.show()
```
